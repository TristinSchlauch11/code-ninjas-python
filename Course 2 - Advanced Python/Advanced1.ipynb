### ***Division***

Lots of times when we divide numbers, our answer isn't a whole number. For example:

````python
print(5/2)
print(1/3)
````

In Python, this decimal number is called a **float**. Floats are different than the integers that we learned about in Intro to Python because they have a decimal point with some numbers after it. This includes the number 0.

````python
print(4/2)
````

There are some other uncommon, but important division operations that you should know. First, there is the **modulo** operator. This operator is written with the percentage sign (%) and returns the remainder after dividing one number by another.

For example, since 9 / 4 = 2 with a remainder of 1, we get that 9 % 4 = 1.

````python
print(9 / 4)
print(9 % 4)
````

Modulo can be used to help determine if something is divisible by something else, what day of the week it is, whose turn it is in a game, and so on. We should see more example of this later.

The other operator is called **floor division**. This operator is written with two slashes (//) and returns the result dividing one number by another *without* the remainder. For example, since 9 / 4 = 2 with a remainder of 1, we get that 9 // 4 = 2.

````python
print(9 / 4)
print(9 // 4)
````

This operation is useful when you are interested in dividing two numbers, but don't need to know the remainder. It's also useful when you want to round the result down. There will be more examples later.

Notice that division *always* returns a float, while floor division and modulo *always* return an integer. This is important for the next section.

### ***Type Conversions***

Sometimes, we are able to convert values between strings, integers, and floats. To do this, there are three **conversion methods** that we use:

```
str() - converts the input to a string
int() - converts the input to an integer
float() - converts the input to a float
```

Below, you can see some examples of converting different values into different types:

````python
print(str(20))      # convert integer to string
print(str(8.5))     # convert float to string
print(int("15"))    # convert string to integer
print(int(4.0))     # convert float to integer
print(float("7.2")) # convert string to float
print(float("8"))   # convert 'integer string' to float
print(float(5))     # convert integer to float
````

But, we need to be a little bit careful when converting between different types.

- When you convert from a float to an integer, you will lose the numbers behind the decimal point

````python
print(int(20.99))
````

- When you convert strings with a decimal to an integer, you will get an error

````python
print(int("3.14"))
````

There are many reasons that we would want to use type conversions.

For example, whenever you use the input statement, the user's answer will *always* be stored as a string. But, if you want the user to input a number, you can use an integer conversion to change the string into an integer.

````python
age = input("How old are you? ")
age = age + 5
print(age)
````

````python
age = input("How old are you? ")
age = int(age)
age = age + 5
print(age)
````

Another reason that we would want to use type conversions is to print numbers and strings together. If we try to combine a string with a number by using the plus sign (+), we will get an error.

````python
print("McDavid wears number " + 97)
````

But, if we convert the number to a string using a type conversion, we are able to combine them!

````python
print("McDavid wears number " + str(97))
````

## **Lists**: Changing List Contents, and the Range Function

### ***Adding, Removing, and Changing Elements***

In Intro to Python, we learned a little bit about how to change the contents of a list by using the append() and remove() functions. Now, we're going to learn a few more ways that we can modify a list.

First, we learned about how we can add a new element to the end of a list by using the append() function

````python
letter_list = ['a', 'b', 'c']
letter_list.append('d')
print(letter_list)
````

We can also add elements to *specific locations* in a list by using the **insert()** function. This function uses two parameters:

- the index of the location of where you want to add the new element, and
- the element you want to add

Remember that computers start counting at 0! So, if you want to add the element to the 2nd spot, you need to add it at index 1.

````python
letter_list.insert(1, 'e')
print(letter_list)
````

We can also add another entire list to the end of our list by using the **extend()** function. This function uses only one parameter, the list that we want to add to the end of our original list. For example:

````python
letter_list.extend(['f', 'g', 'h', 'i'])
print(letter_list)
````

We can also combine two lists together by using the plus sign between the two lists! However, if we want to use the result of combining these two lists, we need to assign the result to a variable.

````python
letter_list = ['j', 'k', 'l'] + letter_list
print(letter_list)
````

Our list is getting pretty big. Let's move on to removing items. In Intro to Python, we learned about how to remove specific elements by using the remove() function.

````python
letter_list.remove('e')
print(letter_list)
````

Another way that we can remove elements from a list is by using the **pop()** function. This function takes up to one parameter, the index of the value that we want to remove from the list.

What makes the pop() function different from the remove() function is that it will *return* the value that is removed from the list. We can assign this value to a variable so that it can be used to do something else in our code!

````python
letter = letter_list.pop(3)
if letter == "a":
  print("We removed 'a'!")
else:
  print("We removed something else...")

print(letter_list.pop(1))
print(letter_list)
````

If you do not include an index to remove, the pop() function will just remove and return the last element in the list.

````python
print(letter_list.pop())
print(letter_list)
````

Finally, do you remember how we learned to get different list elements in Intro to Python? We used square brackets and the index of the element that we wanted to get.

````python
print(letter_list[2])
````

We can also use this format to update the contents of a list! To update the contents of a list at a specific index, we use the format:

```
list_name[index] = new_elem
```

This will change the element at the provided index to be the new element, 'new_elem'. For example:

````python
letter_list[2] = 'z'
print(letter_list[2])
print(letter_list)
````

### ***Range Functions***

Time to learn about something completely new! A **range function** can be used to generate a list of consecutive numbers, normally used in a for loop to run the loop a specific number of times. There are three types of range functions.

1. range(stop) - generates a list of numbers from 0 up to (but not including) the stopping number

````python
list(range(10))
````

2. range(start, stop) - generates a list of numbers from the starting number up to (but not including) the stopping number

````python
list(range(4, 10))
````

3. range(start, stop, step) - generates a list of numbers from the starting number, increasing by the stepping number each time, up to (but not including) the stopping number

````python
list(range(4, 10, 2))
````

Normally, we won't use the 3rd version very often, but we do use the 1st version quite often in for loops. Instead of a list, we'll use a range in our for loop statement. For example:

````python
for i in range(5):
  print(i + i * i)
````

It is also very common to put the length of a list inside of a range function in a for loop:

````python
for i in range(len(letter_list)):
  print(letter_list[i])
````
